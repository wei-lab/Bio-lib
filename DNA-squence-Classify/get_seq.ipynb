```python
import pandas as pd
import numpy as np
import tensorflow as tf
import h5py
import gzip
import os
import random
```

```python
df = pd.read_csv('localtion_seq_dataset.csv',sep=',',header=0)
print df
```

```python
df_index = pd.read_csv('chromosome.txt',sep='\t',header=0)
print df_index
```

```python
dic_seq = {
    'A':[1,0,0,0],
    'C':[0,1,0,0],
    'G':[0,0,1,0],
    'T':[0,0,0,1],
    'a':[1,0,0,0],
    'c':[0,1,0,0],
    'g':[0,0,1,0],
    't':[0,0,0,1],
    'N':[0,0,0,0]}

dic_class = {
    'YES_high':[0,0,1],
    'YES_low':[1,0,0],
    'NO':[0,1,0]
}
```

```python
data_length = 20
```

```python
devided_seq = []
for i in range(len(df_index)):
    filename = ''.join(["./genomes/hg18.fa.split/hg18.id_",df_index.iloc[i].chrom,".fa"])
    replace_str =">"+df_index.iloc[i].chrom
    with open(filename,'r') as mata_data:
        mata_data.readline
        seq = (mata_data.read().replace(replace_str,'')).replace('\n','')
        print len(seq)
        for j in range(len(df)):
            if df.iloc[j].chrom == df_index.iloc[i].chrom:
                inuslator_seq = seq[df.iloc[j].start:df.iloc[j].end]
                for n in range(len(insulator_seq)-data_length):
                    if df.iloc[j].labeled == 'YES' and df.iloc[j].activity < 0.9:
                        devided_seq.append([inuslator_seq[n:n+data_length],'YES_low'])
                    elif df.iloc[j].labeled == 'YES' and df.iloc[j].activity > 0.9:
                        devided_seq.append([inuslator_seq[n:n+data_length],'YES_high'])
                    else:
                        devided_seq.append([inuslator_seq[n:n+data_length],'NO'])
```

```python
final_matix = []
for m in range(len(devided_seq)):
    tmp_matix = []
    if(len(devided_seq[m][0])==20):
        for n in range(len(devided_seq[m][0])):
            tmp_matix.append(dic_seq[str(devided_seq[m][0][n])])
        final_matix.append([tmp_matix,dic_class[str(devided_seq[m][1])]])
print 'Done!!!'
```

```python
labeled = [le[0] for le in final_matix]
classes = [le[1] for le in final_matix]
```

```python
print classes
```

```python
slice_data = random.sample(classes,900)
print slice_data
```

```python
f = h5py.File('localtion_seq_dataset.h5','w')
f.create_dataset('labeled',data=labeled)
f.create_dataset('classes',data=classes)
f.close()
```
